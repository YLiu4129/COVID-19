```python
# Build random forest model and prediction
```

```python
date = pd.to_datetime(train_data['Date'])
```

```python
train_data['Date']= date.dt.strftime("%m%d").astype(int)
```

```python
le = sklearn.preprocessing.LabelEncoder()
```

```python
train_data['Target'] = le.fit_transform(train_data['Target'])
```

```python
feature_cols = ['Population', 'Weight', 'Date', 'Target']
X = train_data[feature_cols] 
y = train_data['TargetValue'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=8787)
```

```python
model = RandomForestRegressor(n_jobs=-1, n_estimators = 50)
model.fit(X_train, y_train)
```

```python
test_date = pd.to_datetime(test_data['Date'])
```

```python
test_data['Date']= test_date.dt.strftime("%m%d").astype(int)
```

```python
test_le = sklearn.preprocessing.LabelEncoder()
```

```python
test_data['Target'] = le.fit_transform(test_data['Target'])
```

```python
test = test_data[feature_cols]
```

```python
test
```

```python
prediction = model.predict(test)
```

```python
prediction_lst = [i for i in prediction]
```

```python
test_df = pd.DataFrame({'number': test_data.index, 'Population': test_data['Population'], 'val': prediction_lst})
```

```python
q_05 = test_df.groupby('number')['val'].quantile(q=0.05).reset_index()
q_50 = test_df.groupby('number')['val'].quantile(q=0.5).reset_index()
q_95 = test_df.groupby('number')['val'].quantile(q=0.95).reset_index()
```

```python
q_05
```

```python
q_05.columns=['number','0.05']
q_50.columns=['number','0.5']
q_95.columns=['number','0.95']
```

```python
temp_df = pd.concat([q_05,q_50['0.5'],q_95['0.95']],1)
temp_df['number'] = temp_df['number'] + 1
```

```python
temp_df
```

```python
sub = pd.melt(temp_df, id_vars=['number'], value_vars=['0.05','0.5','0.95'])
```

```python
sub
```

```python
sub['ForecastId_Quantile']=sub['number'].astype(str)+'_'+sub['variable']
```

```python
sub
```

```python
sub['TargetValue']=sub['value']
```

```python
sub=sub[['ForecastId_Quantile','TargetValue']]
sub.reset_index(drop=True,inplace=True)
sub.to_csv("submission.csv",index=False)
sub.head(10)
```
